# MODEL

```python
X = cars.drop(columns='Price')
y = cars['Price']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type']])
```

```python
ct = ColumnTransformer(
    transformers=[
        ('cat', ohe, ['Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type'])
    ],
    remainder='passthrough'
)
```

```python
best_r2 = -np.inf
best_random_state = None

for i in range(1000):

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    
    lr = LinearRegression()
    
    pipeline = make_pipeline(ct, lr)
    
    pipeline.fit(X_train, y_train)
    
    y_pred = pipeline.predict(X_test)
    
    r2 = r2_score(y_test, y_pred)
    
    if r2 > best_r2:
        best_r2 = r2
        best_random_state = i

print(f"Best RÂ² score: {best_r2}")
print(f"Achieved with random state: {best_random_state}")
```
